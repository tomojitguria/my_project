```python
# Why do we use (n-1) in sample statistics formula whereas for population we use (n)
# Bessel's correction uses n − 1 for correcting the bias in the estimation of the population variance.
```

```python
# Why do we have variance and standard deviation both, when it is almost the same (in layman’s term) just a minor difference of square root.
# 1.Variance is calculated as the average squared deviation of each number from the mean of a data set.
# 2.Standard deviation looks at how spread out data is from the mean, by looking at the square root of the variance.
# 3.Variance helps to find the distribution of data in a population from a mean, and standard deviation also helps to know the distribution of data in population, 
# but standard deviation gives more clarity about the deviation of data from a mean.
```
